```python
import os
from google.generativeai import GenerativeModel, configure
```

```python

configure(api_key=os.getenv("GOOGLE_API_KEY"))
print(os.getenv("GOOGLE_API_KEY"))
```

```python
def query_gemini(question, history=[]):
    model = GenerativeModel("gemini-1.5-pro")
    chat = model.start_chat(history=history)
    response = chat.send_message(question, stream=True)
    response.resolve()  # Ensure response is fully resolved
    return response, chat.history
```

```python
query="create a list of dictionaries with items and its corresponding columns names(fill '0' incase any column is missing for any item). no extra texts or code must be generated."
question = " ".join(str(dfs)) + "\n\n" + query
response, chat_history = query_gemini(question, history=[])

response_text = ""
for chunk in response:
    response_text += chunk.text
print(response_text)
```

```python
data_list = eval(response_text)
data_list
```

```python
df = pd.DataFrame(data_list)
df['Subtotal'] = pd.to_numeric(df['Subtotal'], errors='coerce')
df['Discount (5%)'] = pd.to_numeric(df['Discount (5%)'], errors='coerce')
df['Tax (8%)'] = pd.to_numeric(df['Tax (8%)'], errors='coerce')
df['Tax (10%)'] = pd.to_numeric(df['Tax (10%)'], errors='coerce')
df['Totalinvoice'] = df['Subtotal'] - df['Discount (5%)'] + df['Tax (8%)'] + df['Tax (10%)']
df
```

```python
excel=pd.read_excel(r"Master data.xlsx")
numeric_columns = ['Quantity', 'Unit Price', 'Subtotal', 'Discount (5%)', 'Tax (8%)', 'Tax (10%)', 'Total']
for column in numeric_columns:
    excel[column] = pd.to_numeric(excel[column], errors='coerce')

excel['Totalmaster'] = excel['Subtotal'] - excel['Discount (5%)'] + excel['Tax (8%)'] + excel['Tax (10%)']
excel['Item'] = excel['Item'].replace('Scredriver', 'Screwdriver')
excel
```

```python
newdf=pd.merge(df[['Item', 'Totalinvoice']],
                     excel[['Item', 'Totalmaster']],
                     on='Item',
                     how='left')
newdf.fillna(0,inplace=True)
```

```python
newdf['Discrepancies']=newdf["Totalinvoice"]-newdf["Totalmaster"]
newdf
```
